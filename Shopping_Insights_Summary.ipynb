# Executive Summary: Online Shopper Intention & Conversion Optimization
---

## 1. Dataset Summary & Feature Descriptions

The dataset contains **12,205 unique sessions** (processed after removing 125 duplicate rows) tracking user interactions across 17 distinct behavioral, temporal, and technical features. Here is a short summary description of each feature used:

### Behavioral Features (Page-level Activity)
* **`Administrative` / `Administrative_Duration`**: Number of administrative pages (e.g., account profile, settings) visited and total time spent.
* **`Informational` / `Informational_Duration`**: Number of informational pages (e.g., About Us, contact info) visited and total time spent.
* **`ProductRelated` / `ProductRelated_Duration`**: Number of product pages visited and total time spent; typically forms the core of the consumer browse journey.
* **`BounceRates`**: Percentage of sessions where a user enters a page and leaves immediately without further interactions.
* **`ExitRates`**: Percentage of views to a page that were the final session exit point across the site.
* **`PageValues`**: The average monetary value attributed to a webpage based on its historical presence in successful checkout funnels. **(Highest predictor of intent)**.

### Temporal & Situational Context
* **`SpecialDay`**: Proximity score (0.0 to 1.0) indicating how close the visit date is to major high-volume holidays/special events (e.g., Mother's Day, Valentine's).
* **`Month`**: Calendar month of the session.
* **`Weekend`**: Binary flag (True/False) showing if the visit occurred over the weekend.

### User Profile & Technical Infrastructure
* **`OperatingSystems` / `Browser`**: Identifiers for the user's OS platform and web browser software.
* **`Region`**: Geographic location identifier of the visitor.
* **`TrafficType`**: The specific channel or source through which the user arrived at the site (e.g., paid search, direct, organic).
* **`VisitorType`**: Categorization of users into *Returning_Visitor*, *New_Visitor*, or *Other*.

### Target Variable
* **`Revenue`**: The primary target column indicating transaction success (`1` for purchase, `0` for no transaction).

## 2. Model Evaluation & Performance Ranking

Since the target variable exhibits a heavy class imbalance (~15.6% conversions), **F1-Score** and **ROC-AUC** serve as our primary criteria for model effectiveness, balancing the trade-off between Precision and Recall.

### Model Performance Metrics

| Rank | Model | F1 Score | ROC-AUC | Accuracy | Precision | Recall |
| :---: | :--- | :---: | :---: | :---: | :---: | :---: |
| **1** | **AdaBoost** | **0.6703** | 0.8574 | 87.71% | 0.5777 | 0.7984 |
| **2** | **Random Forest** | **0.6525** | 0.9098 | 85.95% | 0.5322 | 0.8429 |
| **3** | **Decision Tree** | **0.6388** | **0.9196** | 84.76% | 0.5077 | 0.8613 |
| **4** | **Logistic Regression** | **0.6245** | 0.9108 | 85.17% | 0.5172 | 0.7880 |
| **5** | **Support Vector Machine (SVM)** | **0.3464** | 0.8403 | 43.26% | 0.2113 | **0.9607** |

### Performance Analysis & Insights
* **Best model - AdaBoost:** Achieved the highest F1-Score (`0.6703`) and overall accuracy (`87.71%`), making it the most well-rounded option for operational deployment.
* **Strong Tree Contenders:** `Decision Tree` achieved the highest absolute **ROC-AUC (0.9196)**, showing exceptionally strong class discriminative capability.
* **SVM Failure Mode:** The unscaled, un-encoded distance calculations caused SVM to collapse toward guessing or assigning almost everything aggressively to match high-recall behaviors, leading to a bottom-tier accuracy of `43.26%`.

## 3. Key Insights & Findings

* **The 'Page Value' Multiplier:** `PageValues` emerged as the single most powerful predictive driver of intent. High page values correlate exponentially with customer conversion.
* **High Friction Alerts:** Spikes in `BounceRates` and `ExitRates` on primary browse categories signal interface design issues or unexpected price changes.
* **The Imbalance Challenge:** Standard architectures tend to favor the majority non-converting class unless thresholds are dynamically shifted or explicitly weighted.

## 4. Strategic Business Recommendations

1. **Dynamic Target Prompts:** Leverage the high predictive weight of `PageValues` by triggering interactive assistant pop-ups or loyalty offers the moment a visitor hits key validation pages.
2. **Cart Abandonment and Retargeting:** Optimize targeted exit-intent promotions for *Returning Visitors* who drop off near payment options.
3. **Friction Audits:** Streamline UI components on individual landing paths experiencing high `ExitRates` by adding upfront pricing transparently and simplified guest checkout configurations.
4. **Threshold Shift for Production:** Deploy the **AdaBoost** model with tuned probability thresholds or adjusted class-weights to actively capture true buyers without letting high organic non-conversions skew inferences.
